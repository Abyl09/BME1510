#2) What is the average year in program for the class? Master‚Äôs students? PhD students?

```python

#2.1) take average of class data
#Class_mean = Lecture_1.mean()
#you recieve an error due to the invalid enteries 

#print all unique values in the year in rograms columns to see what we are working with
print(Lecture_1['Year_in_program'].unique())

#values needs to be treated as string, but then convert then to numeric (this needs to be done to perform mean function)
#errors='coerce': Converts invalid entries to NaN
Lecture_1['Year_in_program'] = pd.to_numeric(Lecture_1['Year_in_program'], errors = 'coerce')

#filter for all students to include values between 1 and 5
Class_year_filtered = Lecture_1.loc[Lecture_1['Year_in_program'].between(1, 5, inclusive = 'both'), 'Year_in_program']

#find average year using mean for year in program for the class
Class_average = Class_year_filtered.mean()
print('The average year in program is:', Class_average)
```

```python
#2.2) take average year in Master's program for the class
#print all unique values in the Program column to see what we are working with
print(Lecture_1['Program'].unique())

#filter for Master's program only and the year in program for that program
Masters_year_data = Lecture_1.loc[Lecture_1['Program'].isin(['Masters', 'MEng', 'MASc', 'meng', 'Meng', 'MAsc', 'MEng ', 'MASc ', 'MSc ']), ['Program', 'Year_in_program']]

#print all unique values in the year in programs to see what we are working with
print(Masters_year_data['Year_in_program'].unique())

#values needs to be treated as string, but then convert then to numeric (this needs to be done to perform mean function)
#errors='coerce': Converts invalid entries to NaN
Masters_year_data['Year_in_program'] = pd.to_numeric(Masters_year_data['Year_in_program'], errors = 'coerce')

#filter for all students to include values between 1 and 5
Masters_year_filtered = Masters_year_data.loc[Lecture_1['Year_in_program'].between(1, 5, inclusive = 'both'), 'Year_in_program']

#find average year using mean for average year in master's program
Masters_average = Masters_year_filtered.mean()
print('The average year in the Master\'s program is:', Masters_average)
```

```python
#2.3) take average year in PhD program for the class
#print all unique values in the Program column to see what we are working with
print(Lecture_1['Program'].unique())

#filter for PhD program only and the year in program for that program
Phd_year_data = Lecture_1.loc[Lecture_1['Program'].isin(['PhD', 'Phd']), ['Program', 'Year_in_program']]
print(len(Phd_year_data))

#print all unique values in the year in programs to see what we are working with
print(Phd_year_data['Year_in_program'].unique())

#values needs to be treated as string, but then convert then to numeric (this needs to be done to perform mean function)
#errors='coerce': Converts invalid entries to NaN
Phd_year_data['Year_in_program'] = pd.to_numeric(Phd_year_data['Year_in_program'], errors = 'coerce')

#filter for all students to include values between 1 and 5
Phd_year_filtered = Phd_year_data.loc[Lecture_1['Year_in_program'].between(1, 5, inclusive = 'both'), 'Year_in_program']

#find average year using mean for average year in phd program
Phd_average = Phd_year_filtered.mean()
print('The average year in the PhD program is:', Phd_average)
```

#3) how many/what percentage/what proportion of students have experience with Python?

```python
#print all unique values in the coding experience column to see what we are working with
print(Lecture_1['Python_experience'].unique())

#filter for yes and some Python experience column
Python_experience_data = Lecture_1.loc[Lecture_1['Python_experience'].isin(['yes', 'Yes', 'y', 'yep', 'yes ', 'YES']), ['Python_experience']]

#find total count of how many students have coding experience
Total_python_experience = len(Python_experience_data)
print(Total_python_experience, 'student have experience with Python')

#find proportion of how many students have experience with Python
Python_experience_proportion = Total_python_experience / len(Lecture_1)
print('The proportion of students that have experience with Python is:', Python_experience_proportion)

#find percentage of how many students have experience with Python
Python_experience_percentage = Python_experience_proportion * 100
print('The percentage of students that have experience with Python is:', Python_experience_percentage)
```

#4) What proportion of students in a Master‚Äôs program have experience with Python?

```python
#filter for Master's program and python experience column
Masters_python_data = Lecture_1.loc[Lecture_1['Program'].isin(['Masters', 'MEng', 'MASc', 'meng', 'Meng', 'MAsc', 'MEng ', 'MASc ', 'MSc ']), ['Program', 'Python_experience']]

#filter for Masters student with yes/some python experience 
Masters_and_python_exp = Masters_python_data.loc[Masters_python_data['Python_experience'].isin(['yes', 'Yes', 'y', 'yep', 'yes ', 'YES'])]
print(Masters_and_python_exp)

#find total count of masters students with python experience
total_masters_and_python_exp = len(Masters_and_python_exp)
print(total_masters_and_python_exp, 'student are in a Master\'s program and have Python experience')

#find proportion of how many studens are in a masters program and have python experience
Masters_and_python_exp_proportion = (total_masters_and_python_exp) / len(Masters_python_data)
print('The proportion of students that are in a Master\'s program ad have experience with Python is:', Masters_and_python_exp_proportion)
```
